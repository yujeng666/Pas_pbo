```python
import pandas as pd

path ='/content/drive/MyDrive/CSV/indonesia - 3.csv'
df = pd.read_csv(path)
df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='Life expectancy', y='percentage expenditure')

plt.title('table indonesia')
plt.xlabel('Life expectacy')
plt.ylabel('percentage expenditure')
plt.xlim(66, 70)
plt.ylim(-1, 15)
plt.grid(True)
plt.show()
```

```python
import numpy as np
X = np.array(df['Life expectancy']) 
y = np.array(df['percentage expenditure']) 
print(f'X : {X}') 
print(f'y : {y}')
```

```python
X = X.reshape(-1, 1)
X.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() 

model.fit(X, y)
```

```python
X_vis = np.array([0, 10000]).reshape(-1, 1)
y_vis = model.predict(X_vis) 
```

```python
plt.scatter(X, y)
plt.plot(X_vis, y_vis,'-r')
plt.title('table indonesia')
plt.xlabel('Life expectacy')
plt.ylabel('percentage expenditure')
plt.xlim(66, 70)
plt.ylim(-1, 15)
plt.grid(True)
plt.show()
```

```python
print(f'intercept : {model.intercept_}')
print(f'slope : {model.coef_}')
```

```python
life = np.array([23, 52, 32, 12]).reshape(-1, 1)
life
```

```python
prediksi_indeks = model.predict(life)
prediksi_indeks
```

```python
for life, precentace in zip (life, prediksi_indeks) : 
  print(f'life  : {life} precentace : {precentace}')
```
